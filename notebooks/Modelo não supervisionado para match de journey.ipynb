# Treinamento do modelo


```python
# Separar o X do treinamento (input)

X = users[['superfit_dis', 'superfit_sin', 'superfit_cur',
       'superfit_int', 'superfit_eng', 'superfit_res', 'interest',
       'lifestyle_classic', 'lifestyle_order', 'lifestyle_change',
       'lifestyle_tireless', 'lifestyle_explorer', 'lifestyle_specialist',
       'lifestyle_generalist', 'lifestyle_hybrid', 'objective']]
```

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler

# Padronizar a matriz de features
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# Definir o KNN com 5 vizinhos e usando a distancia euclidiana
knn = NearestNeighbors(n_neighbors=5, metric='euclidean')

# Treinar o modelo à matriz de características padronizadas.
knn.fit(X_std)
```

```python
from sklearn.metrics import silhouette_score

# Usar k-NN para achar o vizinhos para cada ponto
distances, indices = knn.kneighbors(X_std)

# Vamos pegar o vizinho mais próximo (excluindo o próprio ponto) para cada ponto como seu rótulo de cluster.
labels = indices[:, 1]

# Calcular Silhouette Score
score = silhouette_score(X_std, labels)
print("Silhouette Score:", score)
```

O silhouette_score é uma métrica usada para avaliar a qualidade de um agrupamento de dados. Ela mede o quão bem os objetos em cada grupo estão agrupados e separados dos objetos em outros grupos, fornecendo uma pontuação entre -1 e 1. Quanto maior a pontuação, melhor é a qualidade do agrupamento.


```python
import numpy as np

# Calcule a soma das distâncias ao quadrado para o vizinho mais próximo de cada ponto.
wss = np.sum([distances[i][1]**2 for i in range(len(distances))])
print("Within-cluster Sum of Squares:", wss)
```

```python
# Calcule a distância média para o vizinho mais próximo de cada ponto.
avg_distance = np.mean([distances[i][1] for i in range(len(distances))])
print("Average Distance to Nearest Neighbors:", avg_distance)
```

```python
# Classe superfit que vai receber o dados de superfit e tratados

class Superfit():
	def __init__(self, superfit_dis: int, superfit_sin: int, superfit_cur: int, superfit_int: int, superfit_eng: int, superfit_res: int):

		# Faixa de valores válidos
		range = [0, 100]

		self.superfit_dis = superfit_dis
		self.superfit_sin = superfit_sin
		self.superfit_cur = superfit_cur
		self.superfit_int = superfit_int
		self.superfit_eng = superfit_eng
		self.superfit_res = superfit_res


		# Mostrar a mensagem de aviso caso algum dos valores estiver fora do intervalo
		if self.superfit_dis < range[0] or self.superfit_dis > range[1]:
			raise ValueError(f'superfit_dis must be between {range[0]} and {range[1]}')

		if self.superfit_sin < range[0] or self.superfit_sin > range[1]:
			raise ValueError(f'superfit_sin must be between {range[0]} and {range[1]}')

		if self.superfit_cur < range[0] or self.superfit_cur > range[1]:
			raise ValueError(f'superfit_cur must be between {range[0]} and {range[1]}')

		if self.superfit_int < range[0] or self.superfit_int > range[1]:
			raise ValueError(f'superfit_int must be between {range[0]} and {range[1]}')

		if self.superfit_eng < range[0] or self.superfit_eng > range[1]:
			raise ValueError(f'superfit_eng must be between {range[0]} and {range[1]}')

		if self.superfit_res < range[0] or self.superfit_res > range[1]:
			raise ValueError(f'superfit_res must be between {range[0]} and {range[1]}')

		# Ajustar as notas a escala certa
		self.superfit_scaled = superfit_scaler.transform([[self.superfit_dis, self.superfit_sin, self.superfit_cur,
	   self.superfit_int, self.superfit_eng, self.superfit_res]])
```

```python
# Classe lifestyle que vai receber o dados de lifestyle e tratados

class Lifestyle():
	def __init__(self, lifestyle_classic: float, lifestyle_order: float, lifestyle_change: float, lifestyle_tireless: float, lifestyle_explorer: float, lifestyle_specialist: float, lifestyle_generalist: float, lifestyle_hybrid: float):
		range = [0.0, 25.0]

		self.lifestyle_classic = lifestyle_classic
		self.lifestyle_order = lifestyle_order
		self.lifestyle_change = lifestyle_change
		self.lifestyle_tireless = lifestyle_tireless
		self.lifestyle_explorer = lifestyle_explorer
		self.lifestyle_specialist = lifestyle_specialist
		self.lifestyle_generalist = lifestyle_generalist
		self.lifestyle_hybrid = lifestyle_hybrid


		# Mostrar a mensagem de aviso caso algum dos valores estiver fora do intervalo.
		if self.lifestyle_classic < range[0] or self.lifestyle_classic > range[1]:
			raise ValueError(f'lifestyle_classic must be between {range[0]} and {range[1]}')

		if self.lifestyle_order < range[0] or self.lifestyle_order > range[1]:
			raise ValueError(f'lifestyle_order must be between {range[0]} and {range[1]}')

		if self.lifestyle_change < range[0] or self.lifestyle_change > range[1]:
			raise ValueError(f'lifestyle_change must be between {range[0]} and {range[1]}')

		if self.lifestyle_tireless < range[0] or self.lifestyle_tireless > range[1]:
			raise ValueError(f'lifestyle_tireless must be between {range[0]} and {range[1]}')

		if self.lifestyle_explorer < range[0] or self.lifestyle_explorer > range[1]:
			raise ValueError(f'lifestyle_explorer must be between {range[0]} and {range[1]}')

		if self.lifestyle_specialist < range[0] or self.lifestyle_specialist > range[1]:
			raise ValueError(f'lifestyle_specialist must be between {range[0]} and {range[1]}')

		if self.lifestyle_generalist < range[0] or self.lifestyle_generalist > range[1]:
			raise ValueError(f'lifestyle_generalist must be between {range[0]} and {range[1]}')

		if self.lifestyle_hybrid < range[0] or self.lifestyle_hybrid > range[1]:
			raise ValueError(f'lifestyle_hybrid must be between {range[0]} and {range[1]}')

		# Ajustar as notas a escala certa
		self.lifestyle_scaled = lifestyle_scaler.transform([[self.lifestyle_classic, self.lifestyle_order, self.lifestyle_change,
	   self.lifestyle_tireless, self.lifestyle_explorer, self.lifestyle_specialist,
	   self.lifestyle_generalist, self.lifestyle_hybrid]])
```

```python
# Classe Extra que vai receber objetivos e interesses

objectives = ['Procurar emprego', 'Melhorar minha carreira', 'Autoconhecimento', 'Desenvolvimento Pessoal']

interests = ['Finanças / Contabilidade', 'Logistica / Supply Chain', 'Engenharia', 'Tecnologia', 'Produção / Operações',
             'Comercial / Vendas', 'RH', 'Marketing / Comunicação', 'Administrativo', 'Saúde', 'Jurídico']

class Extras():
	# Se o objetivo ou interesse não for achado aparece uma aviso
	def __init__(self, interest: str, objective: str):
		if interest not in interests:
			raise Exception('Interest not found')

		if objective not in objectives:
			raise Exception('Objective not found')

		self.interest = interest
		self.objective = objective

		# Ajustar as notas a escala certa
		self.interest = interest_le.transform([self.interest])[0]
		self.objective = objective_le.transform([self.objective])[0]
```

```python
# Função para gerar a recomendação de journey (curso) para um usuário não existente

def generate_journeys_unknown(superfit: Superfit, lifestyle: Lifestyle, extras: Extras):
	# obtain a numpy array for knn input in the following order:
	# ['superfit_dis', 'superfit_sin', 'superfit_cur',
    #   'superfit_int', 'superfit_eng', 'superfit_res', 'interest',
    #   'lifestyle_classic', 'lifestyle_order', 'lifestyle_change',
    #   'lifestyle_tireless', 'lifestyle_explorer', 'lifestyle_specialist',
    #   'lifestyle_generalist', 'lifestyle_hybrid', 'objective']
	print('Generating journeys...')
	input = np.array([superfit.superfit_dis, superfit.superfit_sin, superfit.superfit_cur,
				   	superfit.superfit_int, superfit.superfit_eng, superfit.superfit_res, extras.interest,
	   				lifestyle.lifestyle_classic, lifestyle.lifestyle_order, lifestyle.lifestyle_change,
	   				lifestyle.lifestyle_tireless, lifestyle.lifestyle_explorer, lifestyle.lifestyle_specialist,
	   				lifestyle.lifestyle_generalist, lifestyle.lifestyle_hybrid, extras.objective])

	# Padronizar os inputs
	input_std = scaler.transform(input.reshape(1, -1))

	# Achar os vizihos mais proximos
	distances, indices = knn.kneighbors(input_std)

	# Pegar os ids dos gestores similares
	ids = users.iloc[indices[0]].user_id.values

	# Encontre as jornadas concluídas por esses usuários semelhantes
	similar_user_journeys = user_journeys[user_journeys['user_id'].isin(ids)]

	# Achar as journeys no dataframe
	locate_journeys = journeys.rename(columns={'id': 'journey_id'})

	similar_user_journeys = pd.merge(similar_user_journeys, locate_journeys, on='journey_id', how='left')

	# Recomende as 3 jornadas com as maiores avaliações e também com o maior número de avaliações
	recommended_journeys = similar_user_journeys.sort_values(by=['average_rating', 'ratings_count'], ascending=False)

	# Remover duplicadas baseados no journey_id
	recommended_journeys.drop_duplicates(subset='journey_id', inplace=True)

	# Dropar a coluna de user_id
	recommended_journeys.drop('user_id', axis=1, inplace=True)

	return recommended_journeys.head(3)
```

```python
# Função para gerar a recomendação de journey (curso) para um usuário existente

def generate_journeys(user_id):
	print('Generating journeys...')
	# Achar os vizinhos mais proximos
	given_user = users[users['user_id'] == user_id]

	if given_user.empty:
		raise Exception('User not found')

	given_user = given_user.drop('user_id', axis=1)

	# Padroniazar os inptuts
	input_std = scaler.transform(given_user)

	# Achar os vizinhos mais proximos
	distances, indices = knn.kneighbors(input_std)

	# Pegar os ids dos usuários
	ids = users.iloc[indices[0]].user_id.values

	# Encontrar as jornadas completadas por esses usuários similares
	similar_user_journeys = user_journeys[user_journeys['user_id'].isin(ids)]

	# Remover todas as viagens que o usuário já completou
	similar_user_journeys = similar_user_journeys[~similar_user_journeys['journey_id'].isin(user_journeys[user_journeys['user_id'] == user_id].journey_id)]

	# Encontrar aquelas viagens no dataframe de viagens
	locate_journeys = journeys.rename(columns={'id': 'journey_id'})

	similar_user_journeys = pd.merge(similar_user_journeys, locate_journeys, on='journey_id', how='left')

	# Recomendar as 3 jornadas mais bem avaliadas e com maior contagem de avaliações.
	recommended_journeys = similar_user_journeys.sort_values(by=['average_rating', 'ratings_count'], ascending=False)

	# Remover duplicatas baseadas na coluna de journey_id
	recommended_journeys.drop_duplicates(subset='journey_id', inplace=True)

	# Dropar a coluna de user_id
	recommended_journeys.drop('user_id', axis=1, inplace=True)

	# Retornar apenas top 3 recomendações
	return recommended_journeys.head(3)
```
