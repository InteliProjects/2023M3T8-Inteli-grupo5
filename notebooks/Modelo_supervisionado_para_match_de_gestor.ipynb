##Média dos Superfit - Candidato x Vagas

---

````python
# Valores das colybas
u_scores = df_user_superfit[['superfit_dis',	'superfit_sin',	'superfit_cur',	'superfit_int',	'superfit_eng',	'superfit_res']]
j_scores = df_job_superfit[['score_res', 'score_eng', 'score_int', 'score_cur', 'score_sin', 'score_dis']]
````

````python
# Reorganizar colunas para ficar compativel com a j_scores
u_scores = u_scores[['superfit_res',	'superfit_eng',	'superfit_int',	'superfit_cur',	'superfit_sin',	'superfit_dis']]

u_scores
````

````python
# Padronizar as colunas do scores dos usuários
u_scores.columns = j_scores.columns

# Média das colunas
u_mean = u_scores.mean()
j_mean = j_scores.mean()

# Criar um DataFrame com as médias
mean_df = pd.DataFrame({'u_superfit_mean': u_mean, 'j_superfit_mean': j_mean})

# Transpor o DataFrame para que as médias sejam uma única linha
mean_df = mean_df.T

mean_df
````

````python
# Dados dos scores
scores_vagas = mean_df.iloc[1,0:].values # vagas
scores2_usuários = mean_df.iloc[0,0:].values # usuários

# Categorias correspondentes aos scores
categories = ['score_res', 'score_eng', 'score_int', 'score_cur', 'score_sin', 'score_dis']

# Preparando o ângulo para cada categoria
num_categories = len(categories)
angles = np.linspace(0, 2 * np.pi, num_categories, endpoint=False)

# Fazendo o gráfico de radar
fig, ax = plt.subplots(figsize=(5, 5), subplot_kw={'polar': True})

# Plot dos valores das vagas
ax.plot(angles, scores_vagas, 'o-', linewidth=2, label='Superfit Vagas')
ax.fill(angles, scores_vagas, alpha=0.25, color='blue')  # Preenchimento da área entre os pontos e o centro

# Plot dos valores do user
ax.plot(angles, scores2_usuários, 'o-', linewidth=2, label='Superfit Usuários')
ax.fill(angles, scores2_usuários, alpha=0.25, color='orange')  # Preenchimento da área entre os pontos e o centro

# Configurações adicionais
ax.set_theta_offset(np.pi/2)  # Começa o gráfico a partir do topo
ax.set_theta_direction(-1)    # Sentido horário
ax.set_rlabel_position(0)      # Posição das labels dos rótulos

# Adicionando labels às categorias
ax.set_xticks(angles)
ax.set_xticklabels(categories)

# Adicionando uma grade radial
ax.yaxis.grid(True)

# Adicionando um título
plt.title('Radar Chart - Média dos superfit - Vagas x Usuários')

# Mostrando o gráfico
plt.show()
````

---

**Análise do Gráfico**



Esse gráfico de radar compara as médias dos scores de superfit de usuários e gestores. A partir disso foi possível observar que a média dos superfit dos usuário é maior que a dos gestores em todas variáveis, em pelo menos 20 pontos. Além disso, esse tipo de gráfico nos possibilita facilitar a análise visual das diferenças nas médias, identificar rapidamente as áreas onde as médias dos dois grupos estão alinhadas e onde há discrepâncias significativas, ajuda a  entender se as médias dos scores estão consistentes ou se variam amplamente.
A análise desse gráfico pode ser feita de diversas maneiras, como por exemplo observando as diferenças e distância das pontuações nas formas dos polígonos entre os grupos, identificando quais scores estão mais alinhados ou divergentes. Além do mais, a compreensão das médias dos scores dos usuários e gestores nos ajuda a entender quais variáveis podem ter uma influência maior no resultado do nosso modelo preditivo e a validar nossas hipóteses.

\## Gráfico da Resultados por Percentil 75 do Superfit

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
coluna_superfit = ['superfit_dis', 'superfit_sin', 'superfit_cur', 'superfit_int', 'superfit_eng', 'superfit_res']
#Pegando os percentil 75
coluna_resultado_v = [64, 56, 50, 58, 56, 58]
coluna_resultado_u = [36, 34, 38, 32, 38, 42]
data = {
    'coluna_superfit': coluna_superfit,
    'coluna_resultado_v': coluna_resultado_v,
    'coluna_resultado_u': coluna_resultado_u
}
df = pd.DataFrame(data)
x = np.arange(len(coluna_superfit))  # Localização dos grupos no eixo x
width = 0.35  # Largura das barras
fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, df['coluna_resultado_u'], width, label='Resultado Vagas')
rects2 = ax.bar(x + width/2, df['coluna_resultado_v'], width, label='Resultado User')
ax.set_xlabel('Superfit')
ax.set_ylabel('Resultado')
ax.set_title('Gráfico de Resultados por Superfit')
ax.set_xticks(x)
ax.set_xticklabels(df['coluna_superfit'])
ax.legend()
# Adicione rótulos nas barras
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # Deslocamento vertical
                    textcoords="offset points",
                    ha='center', va='bottom')
autolabel(rects1)
autolabel(rects2)
fig.tight_layout()
plt.show()
````

---

**Análise do Gráfico**

Analisando o gráfico de superfit, construído com base na média das melhores respostas dos conjuntos de dados disponíveis (percentil 75), é evidente que as competências de 'consciência', 'engajamento' e 'resiliência' demonstram as maiores discrepâncias entre os diferentes resultados. Essas informações podem ser utilizadas para identificar os três cursos mais adequados para um perfil específico. No contexto dessa análise, foram consideradas diversas respostas, porém, caso consigamos aplicar a mesma abordagem de forma individualizada, poderíamos desenvolver um programa capaz de prever quais cursos são mais apropriados para um candidato com base na vaga almejada. Isso abriria a possibilidade de uma orientação personalizada, utilizando-se de estratégias similares, para sugerir as melhores opções de formação.

Pode ser que a competência resiliente não seja uma das principais considerações nas correspondências.

Uma correlação positiva indica que candidatos com habilidades colaborativas podem ter mais correspondências positivas com as vagas.
- **score_cur (Original)**: Correlação fraca e positiva (0.048058). A competência "original" apresenta uma correlação fraca e positiva, indicando que candidatos com essa habilidade podem ter alguma influência nas correspondências, mas não é tão determinante.
- **score_sin (Pragmatico)**: Correlação fraca e positiva (0.017968). A competência "pragmático" dos candidatos também apresenta uma correlação fraca e positiva, indicando que essa habilidade pode ter um papel menos significativo nas correspondências.
- **score_dis (Consistente)**: Correlação moderada e positiva (0.089666). A competência "consistente" dos candidatos apresenta uma correlação moderada e positiva com as características das vagas. Isso sugere que essa habilidade pode ter uma influência mais relevante nas correspondências.

## Obter 3 melhores gestores (que existem) para cada usuario (que existe) utilizando o algoritmo de aproximacao KNN.

````python
# Igualar os df a outras váriaveis
user_data = normalized_df_user
gestor_data = normalized_df_job

# Extrair os IDs de usuários e gestores e dropar alguma colunas que não vamos usar
user_ids = user_data["id"]
user_features = user_data.drop(columns=["id", "lifestyle_hybrid"]).values

gestor_ids = gestor_data["id"]
gestor_features = gestor_data.drop(columns=["id", "job_opportunity_id_y", "job_opportunity_id_x"]).values
````

````python
import numpy as np
from sklearn.neighbors import NearestNeighbors

# Definir o número de vizinhos
k = 3

# Insciar o modelo de KNN
knn = NearestNeighbors(n_neighbors=k, metric='euclidean')  # You can choose a different metric if needed

# Treinar o modelo com os dados dos gestores
knn.fit(gestor_features)

# Achar a distancia e ids dos gestores mais próximos
distances, closest_gestor_indices = knn.kneighbors(user_features, n_neighbors=k)

# Pegar os resultados dos top 3 gestores
results_1 = np.column_stack((user_ids, gestor_ids[closest_gestor_indices[:, 0]]))
results_2 = np.column_stack((user_ids, gestor_ids[closest_gestor_indices[:, 1]]))
results_3 = np.column_stack((user_ids, gestor_ids[closest_gestor_indices[:, 2]]))

distances_1 = np.column_stack(distances[:, 0])

# Printar resultado final
results_final = pd.DataFrame({
    'User_ID': user_ids, 'Gestor 1': results_1[: , 1], 'Distance 1': distances[:, 0], 'Gestor 2': results_2[: , 1], 'Distance 2': distances[:, 1], 'Gestor 3': results_3[: , 1], 'Distance 3': distances[:, 2]
    })

results_final
````

### Função para prever um gestor dado um usuário

Criamos esse função que pode ser utilizada para a implementação dos nossos modelos em um backend de uma aplicação web.

````python
import numpy as np
import pandas as pd
from sklearn.neighbors import NearestNeighbors

def melhores_gestores(user_input, modelo_preditivo, gestor_df, N_vagas):
    # Predições feitas pelo modelo
    predicted = modelo_preditivo.predict(user_input)

    # Tirar as colunas inuteis
    gestor_df = gestor_df.drop(columns=["job_opportunity_id_x", "job_opportunity_id_y"])

    # Criar um DataFrame com os valores previstos e colunas do gestor_df.
    predicted_df = pd.DataFrame(predicted, columns=gestor_df.drop(columns=['job_id']).columns)

    # Inicializar um modelo de NearestNeighbors para encontrar os gestores mais próximos.
    knn = NearestNeighbors(n_neighbors=N_vagas)

    # Ajustar o modelo usando os dados dos gerentes, excluindo a coluna 'job_id'
    knn.fit(gestor_df.drop('job_id', axis=1))

    # Encontre os 3 gerentes mais próximos das preferências do usuário.
    distances, indices = knn.kneighbors(predicted_df)

    # Obter os IDs de trabalho dos gerentes mais próximos.
    nearest_gestors = gestor_df.iloc[indices[0]]['job_id']

    # Print the job_ids and distances to the nearest managers
    print("Gesores mais proximos (job_ids):", ', '.join(map(str, nearest_gestors.tolist())))
    print("Distancias dos gestores mais proximos:", ', '.join(map(str, distances.tolist()[0])) + ", respectively")

    print('\n')

    # Informações completas dos gestores
    nearest_gestor_info = gestor_df[gestor_df['job_id'].isin(nearest_gestors.tolist())]
    print("Information about the nearest gestors:")
    print(nearest_gestor_info.to_string(index=False))
````
