```python
# Pre processamento

# Importar função do sklearn
from sklearn.preprocessing import MinMaxScaler


# Simular um ajuste para o superfit_scaler
dummy_df_superfit = pd.DataFrame({
    'superfit_dis': [0, 100],
    'superfit_sin': [0, 100],
    'superfit_cur': [0, 100],
    'superfit_int': [0, 100],
    'superfit_eng': [0, 100],
    'superfit_res': [0, 100]
})

# Aplicar o MixMaxScaler nos dados de superfit
superfit_scaler = MinMaxScaler()
superfit_scaler.fit(dummy_df_superfit)

# Simular um ajuste para o superfit_scaler
dummy_df_lifestyle = pd.DataFrame({
    'lifestyle_classic': [0, 25],
    'lifestyle_order': [0, 25],
    'lifestyle_change': [0, 25],
    'lifestyle_tireless': [0, 25],
    'lifestyle_explorer': [0, 25],
    'lifestyle_specialist': [0, 25],
    'lifestyle_generalist': [0, 25]
})

# Aplicar o MixMaxScaler nos dados de superfit
lifestyle_scaler = MinMaxScaler()
lifestyle_scaler.fit(dummy_df_lifestyle)
```

# Teste do modelo

```python
# Classe superfit que vai receber o dados de superfit e tratados
class Superfit():
	def __init__(self, superfit_dis: int, superfit_sin: int, superfit_cur: int, superfit_int: int, superfit_eng: int, superfit_res: int):
		range = [0, 100]

		self.superfit_dis = superfit_dis
		self.superfit_sin = superfit_sin
		self.superfit_cur = superfit_cur
		self.superfit_int = superfit_int
		self.superfit_eng = superfit_eng
		self.superfit_res = superfit_res


		# Mostrar a mensagem de aviso caso algum dos valores estiver fora do intervalo
		if self.superfit_dis < range[0] or self.superfit_dis > range[1]:
			raise ValueError(f'superfit_dis must be between {range[0]} and {range[1]}')

		if self.superfit_sin < range[0] or self.superfit_sin > range[1]:
			raise ValueError(f'superfit_sin must be between {range[0]} and {range[1]}')

		if self.superfit_cur < range[0] or self.superfit_cur > range[1]:
			raise ValueError(f'superfit_cur must be between {range[0]} and {range[1]}')

		if self.superfit_int < range[0] or self.superfit_int > range[1]:
			raise ValueError(f'superfit_int must be between {range[0]} and {range[1]}')

		if self.superfit_eng < range[0] or self.superfit_eng > range[1]:
			raise ValueError(f'superfit_eng must be between {range[0]} and {range[1]}')

		if self.superfit_res < range[0] or self.superfit_res > range[1]:
			raise ValueError(f'superfit_res must be between {range[0]} and {range[1]}')

		# Ajustar as notas a escala certa
		self.superfit_scaled = superfit_scaler.transform([[self.superfit_dis, self.superfit_sin, self.superfit_cur,
	   	self.superfit_int, self.superfit_eng, self.superfit_res]])
```

```python
# Classe lifestyle que vai receber o dados de lifestyle e tratados

class Lifestyle():
	def __init__(self, lifestyle_classic: float, lifestyle_order: float, lifestyle_change: float, lifestyle_tireless: float, lifestyle_explorer: float, lifestyle_specialist: float, lifestyle_generalist: float):
		range = [0.0, 25.0]

		self.lifestyle_classic = lifestyle_classic
		self.lifestyle_order = lifestyle_order
		self.lifestyle_change = lifestyle_change
		self.lifestyle_tireless = lifestyle_tireless
		self.lifestyle_explorer = lifestyle_explorer
		self.lifestyle_specialist = lifestyle_specialist
		self.lifestyle_generalist = lifestyle_generalist

		# Mostrar a mensagem de aviso caso algum dos valores estiver fora do intervalo.
		if self.lifestyle_classic < range[0] or self.lifestyle_classic > range[1]:
			raise ValueError(f'lifestyle_classic must be between {range[0]} and {range[1]}')

		if self.lifestyle_order < range[0] or self.lifestyle_order > range[1]:
			raise ValueError(f'lifestyle_order must be between {range[0]} and {range[1]}')

		if self.lifestyle_change < range[0] or self.lifestyle_change > range[1]:
			raise ValueError(f'lifestyle_change must be between {range[0]} and {range[1]}')

		if self.lifestyle_tireless < range[0] or self.lifestyle_tireless > range[1]:
			raise ValueError(f'lifestyle_tireless must be between {range[0]} and {range[1]}')

		if self.lifestyle_explorer < range[0] or self.lifestyle_explorer > range[1]:
			raise ValueError(f'lifestyle_explorer must be between {range[0]} and {range[1]}')

		if self.lifestyle_specialist < range[0] or self.lifestyle_specialist > range[1]:
			raise ValueError(f'lifestyle_specialist must be between {range[0]} and {range[1]}')

		if self.lifestyle_generalist < range[0] or self.lifestyle_generalist > range[1]:
			raise ValueError(f'lifestyle_generalist must be between {range[0]} and {range[1]}')

		# Ajustar as notas a escala certa
		self.lifestyle_scaled = lifestyle_scaler.transform([[
			self.lifestyle_classic, self.lifestyle_order, self.lifestyle_change,
	   		self.lifestyle_tireless, self.lifestyle_explorer, self.lifestyle_specialist,
	   		self.lifestyle_generalist
		]])
```

```python
# Função de teste para gerar os matchs de usuário com gestor

def generate_jobs(superfit: Superfit, lifestyle: Lifestyle):
	print('Generating jobs...')

 # Carregar os inputs do usuários
	input = np.array([
		superfit.superfit_dis, superfit.superfit_sin, superfit.superfit_cur,
		superfit.superfit_int, superfit.superfit_eng, superfit.superfit_res,
		lifestyle.lifestyle_classic, lifestyle.lifestyle_order, lifestyle.lifestyle_change,
		lifestyle.lifestyle_tireless, lifestyle.lifestyle_explorer, lifestyle.lifestyle_specialist,
		lifestyle.lifestyle_generalist
	])

	# Transformar a entrada em um dataframe com nomes de coluna
	input = pd.DataFrame(input.reshape(1, -1), columns=[
		'superfit_dis', 'superfit_sin', 'superfit_cur',
		'superfit_int', 'superfit_eng', 'superfit_res',
		'lifestyle_classic', 'lifestyle_order', 'lifestyle_change',
		'lifestyle_tireless', 'lifestyle_explorer', 'lifestyle_specialist',
		'lifestyle_generalist'
	])

	# Encontrar os vizinhos mais próximos atráves do KNN
	distance, indices = knn.kneighbors(input)

	print(indices)

	# Obter os IDs dos trabalhos mais similares
	ids = jobs.iloc[indices[0]].job_id.values

	# Encontrar os trabalhos existentes para os IDs
	similar_jobs = jobs[jobs['job_id'].isin(ids)].copy()

	# Ordenar por quão similares os trabalhos são ao usuário
	similar_jobs['similarity'] = knn.kneighbors(input)[0][0]

	# Ordenar por similaridade
	similar_jobs.sort_values('similarity', ascending=False, inplace=True)

	# Dropar a coluna de similaridade
	# similar_jobs.drop('similarity', axis=1, inplace=True)

	# Remover duplicatas baseada no job_id
	similar_jobs.drop_duplicates(subset='job_id', inplace=True)

	return similar_jobs.head(3)
```
